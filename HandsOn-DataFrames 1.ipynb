## Drop rows with NaN values

```python
nba = pd.read_csv('pandas/nba.csv')
nba.head(3)
```

```python
nba.tail()
```

```python
# Remove rows where there is a NaN value
nba.dropna().tail()
```

```python
# Drop only values where all rows are NaN
nba.dropna(how='all')
nba.dropna(how='all').tail()
```

```python
nba.dropna(axis='columns')
nba.dropna(axis=1)
```

```python
# Remove a row of NaN from a specific column
nba.dropna(subset=['Salary'])
```

## Fill in Null values with the .fillna() method

```python
nba = pd.read_csv('pandas/nba.csv')
nba.head(3)
```

```python
# Calling fillna on entire DF
nba.fillna(0)
# It will 0 in college column also, that's why it's always better to apply fillna on a series rather than the complete DF
```

```python
nba['Salary'].fillna(0, inplace=True)
nba.head(5)
```

```python
nba['College'].fillna('No college', inplace=True)
nba.head(5)
```

## The .astype() method

```python
nba = pd.read_csv('pandas/nba.csv').dropna(how='all')
nba['Salary'].fillna(0, inplace=True)
nba['College'].fillna('None', inplace = True)
nba.head(5)
```

```python
nba.dtypes
```

```python
nba.info()
```

```python
nba['Salary'] = nba['Salary'].astype('int')
```

```python
nba
```

```python
nba[['Age', 'Weight']] = nba[['Age', 'Weight']].astype('int')
nba
```

```python
# Category data type when there are very less data
nba['Position'].nunique()
nba['Position']=nba['Position'].astype('category')
nba.info()
```

```python
nba['Team'].nunique()
```

```python
nba['Team'] = nba['Team'].astype('category')
nba.info()
```

## Rank values with .rank() method

```python
nba = pd.read_csv('pandas/nba.csv').dropna(how='all')
nba['Salary'] = nba['Salary'].fillna(0).astype('int')
nba.head(3)
```

```python
nba['Salary'].rank(ascending=False).astype('int')
```

```python
nba['Salary Rank']=nba['Salary'].rank(ascending=False).astype('int')
nba
```

```python
nba.sort_values('Salary', ascending=False)
```
